```python
path = "/content/drive/MyDrive/Colab Notebooks/chess price"
```

```python
import tensorflow as tf
```

```python
train_data = tf.keras.preprocessing.image_dataset_from_directory(path,validation_split=0.2,image_size=(224,224),batch_size=32,subset="training",seed=123)
```

```python
test_data = tf.keras.preprocessing.image_dataset_from_directory(path,validation_split=0.2,image_size=(224,224),batch_size=32,subset="validation",seed=123)
```

```python
class_names = ["Queen","Rook","bishop","knight","pawn"]
```

```python
inputs = tf.keras.Input(shape=(224,224,3))
preprocess = tf.keras.applications.mobilenet.preprocess_input(inputs)
upscale = tf.keras.layers.Lambda(lambda x : tf.image.resize_with_pad(x,224,224,method = tf.image.ResizeMethod.BILINEAR))(inputs)
```

```python
from keras.applications import mobilenet
```

```python
mobilenet = mobilenet.MobileNet(include_top="True",weights="imagenet",input_tensor=(upscale),input_shape=(224,224,3))
```

```python
data_agumantation = tf.keras.Sequential([
    tf.keras.layers.experimental.preprocessing.RandomFlip("horizontal"),
    tf.keras.layers.experimental.preprocessing.RandomRotation(0.1),
     tf.keras.layers.experimental.preprocessing.RandomZoom(0.1)
])
```

```python
from tensorflow.keras.layers.experimental.preprocessing import Rescaling
from tensorflow.keras.layers import Flatten,Dense
```

```python
model = tf.keras.Sequential()
model.add(Rescaling(1/255,input_shape=(224,224,3)))
model.add(data_agumantation)
model.add(mobilenet)
model.add(Flatten())
model.add(Dense(128,activation="relu"))
model.add(Dense(len(class_names),activation="softmax"))
```

```python
model.compile(loss=tf.keras.losses.SparseCategoricalCrossentropy(),optimizer="adam",metrics=["accuracy"])
```

```python
model.summary()
```

```python
model.fit(train_data,validation_data=test_data,epochs=10)
```
